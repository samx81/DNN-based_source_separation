## Resume Training

```python
%%shell
export TZ=UTC-9

cd "/content/DNN-based_source_separation/egs/musdb18/mm-dense-lstm"

musdb18_root="../../../dataset/MUSDB18" # Path to MUSDB18 dataset
target='bass' # 'bass', 'drums', 'other', 'vocals'
exp_dir="./exp"
# exp_dir="/content/drive/MyDrive/Colab Notebooks/work/DNN-based_source_separation/egs/musdb18/mm-dense-lstm/exp" for google drive
tag="YourFavoriteTag"
model_path="${exp_dir}/${tag}/model/${target}/last.pth"

config_path="./config/lstm-after-dense/${target}.yaml"
augmentaion_path="./config/paper/augmentation-${target}.yaml"

. ./train.sh \
--musdb18_root "${musdb18_root}" \
--exp_dir "${exp_dir}" \
--tag "${tag}" \
--target "${target}" \
--config_path "${config_path}" \
--augmentaion_path "${augmentaion_path}" \
--continue_from "${model_path}"
```
